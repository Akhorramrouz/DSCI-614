# Q2

### Provide a graphical summary of the frequency of teeth cleaning variable, include some comments on what you see.  

```python
df = pd.read_csv("tootthpaste.txt", sep="\t", index_col=0)

categories = pd.DataFrame(df['Frequency of Teeth Cleaning'].value_counts()).index
counts = pd.DataFrame(df['Frequency of Teeth Cleaning'].value_counts())['Frequency of Teeth Cleaning']

plt.bar(categories,counts)
plt.xticks(rotation=45)
plt.show()
```

<span style='color:blue; font-size:20px'>
    This is a categorical data(it is ordinal) data so the best tool to visualise the data is a bar chart
<span>

### C

Do you think there is a relationship between relationship status and how frequently someone gets their teeth cleaned?  

```python
single = df.loc[df['Single Status'] == "Single"]
not_single = df.loc[~df.index.isin(single.index)]
single_counts = dict(single.drop(columns=['Single Status']).value_counts())
not_single_counts = dict(not_single.drop(columns=['Single Status']).value_counts())
```

```python
categories = [key[0] for key in single_counts.keys()]
values1 = list(single_counts.values())
values2 = list(not_single_counts.values())

# Create an index for the x-axis
x = np.arange(len(categories))
width = 0.35  # Width of the bars

# Create the figure and axis objects
fig, ax = plt.subplots()

# Plot the bars for the first dictionary
bar1 = ax.bar(x - width/2, values1, width, label='Single')
# Plot the bars for the second dictionary
bar2 = ax.bar(x + width/2, values2, width, label='Not Single')

# Set labels, title, and ticks
ax.set_xlabel('Frequency of Teeth Cleaning')
ax.set_ylabel('Count')
ax.set_title('Grouped Bar Chart for Frequency of Teeth Cleaning')
ax.set_xticks(x)
ax.set_xticklabels(categories, rotation=45)
ax.legend()

# Display the chart
plt.tight_layout()
plt.show()
```

<span style="font-size:20px; color:blue">
Based on the graph above those who clean their teeth more frequently are more likley to be in a relationship
<span>
