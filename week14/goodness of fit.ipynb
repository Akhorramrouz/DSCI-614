```python
from scipy.stats import chi2_contingency, chi2

# Observed frequencies
observed_frequencies = [28, 32, 45, 15]

# Expected frequencies (25% of 120 for each department)
expected_frequencies = [120 * 0.25] * 4

# Degrees of freedom
df = len(observed_frequencies) - 1


```

```python
# Perform the chi-square test
chi_square_statistic = sum((o - e)**2 / e for o, e in zip(observed_frequencies, expected_frequencies))
print(chi_square_statistic)
```

```python
critical_value = chi2.ppf(0.95, df)  # 95% confidence level
print(critical_value)  # 95% confidence level)
```

```python
chi_square_statistic, critical_value, chi_square_statistic > critical_value
```

```python
from scipy.stats import chi2
import numpy as np

observed = [926,288,293,104]
expected_prob = np.array([9,3,3,1])/16
```

```python
test_statistic = np.sum([((o-e)**2)/e for o,e in zip(observed,expected_prob*np.sum(observed))])
df = len(observed)-1
print(test_statistic, df)
```

```python
critical_val = chi2.ppf(.95,3)
print(critical_val)
```

```python
p_val = 1 - chi2.cdf(test_statistic, df)
print(p_val)
```
